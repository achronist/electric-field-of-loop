```python
from numpy import *
#import scipy.optimize
import matplotlib.pyplot as plt

from scipy.special import ellipk, ellipe
```

```python
K = lambda k: ellipk(k**2)
E = lambda k: ellipe(k**2)
```

```python
def potential(R,r,z):
    'potential of a loop of unit charge'
    q=(r+R)**2+z**2
    k=sqrt(4*r*R/q)
    return (2/pi)*K(k)/sqrt(q)

def electric(R,r,z):
    'electric field of a loop of unit charge'
    q=(r+R)**2+z**2
    m=4*r*R/q
    k=sqrt(m)
    Kk=K(k)
    Ek=E(k)
    Er=(2/pi)*(1/(q*sqrt(q)*(1-m)*m))*(2*R*Kk*(1-m)-Ek*(2*R-m*(r+R)))
    Ez=(2/pi)*z*Ek/(q*sqrt(q)*(1-m))
    return (Er,Ez)
```

```python
m=mgrid[2:5:.1,2:5:.1]
r=m[0]
z=m[1]
```

```python
R=1
pot=potential(R,r,z)
(Er,Ez)=electric(R,r,z)
```

```python
plt.contour(r,z,pot)
Emag=sqrt(Er**2+Ez**2)
plt.quiver(r,z,Er/Emag,Ez/Emag)
plt.axis('equal')
plt.grid()
plt.show()
```

```python
plt.contour(r,z,Emag*(r**2+z**2),100)
plt.axis('equal')
plt.colorbar()
plt.show()
```
